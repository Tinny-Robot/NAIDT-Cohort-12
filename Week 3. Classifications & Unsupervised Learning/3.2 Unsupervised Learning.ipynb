# Unsupervised Learning


Unsupervised learning is a type of machine learning where the model learns patterns and relationships from unlabeled data without any explicit target variable or predefined output. It aims to discover the underlying structure and patterns within the data. Unsupervised learning algorithms explore the data and identify inherent patterns, clusters, or associations without any prior knowledge.

### Clustering
Clustering is one of the main tasks in unsupervised learning. It involves grouping similar data points together based on their intrinsic properties or characteristics. The goal of clustering is to partition the data into meaningful and homogeneous clusters, where objects within the same cluster are more similar to each other compared to objects in different clusters. Popular clustering algorithms include K-means, Hierarchical Clustering, and DBSCAN.

### Association Rule Mining
Association rule mining is a technique used to discover interesting relationships or associations among items in large datasets. It identifies frequent itemsets and generates rules that express relationships between different items. Association rule mining is commonly used in market basket analysis, where it helps identify patterns in customer purchasing behavior. Apriori and FP-Growth are popular algorithms for association rule mining.

### Dimensionality Reduction
Dimensionality reduction techniques are used to reduce the number of input features while retaining the most important information. This helps in visualizing high-dimensional data and removing redundant or irrelevant features. Dimensionality reduction can be achieved through techniques such as Principal Component Analysis (PCA), t-SNE (t-Distributed Stochastic Neighbor Embedding), and Autoencoders.

### Anomaly Detection
Anomaly detection aims to identify rare or unusual data instances that differ significantly from the majority of the data. It is used to detect outliers or anomalies that do not conform to expected patterns or behaviors. Anomaly detection algorithms can be based on statistical methods, clustering, or supervised learning techniques. Common methods include Isolation Forest, Local Outlier Factor (LOF), and One-Class SVM.

Unsupervised learning techniques are valuable when working with unlabelled data or when the underlying structure or patterns within the data are unknown. They provide insights into the data distribution, help in data exploration, and can be used for various applications such as customer segmentation, anomaly detection, and recommender systems.

Note: Unsupervised learning does not provide a direct measure of accuracy or error, as there are no explicit target variables. Evaluation of unsupervised learning algorithms often relies on qualitative analysis, visualizations, or domain knowledge.

#### K-means Clustering
K-means is a simple yet effective clustering algorithm that partitions the data into K distinct clusters based on their proximity. The algorithm follows an iterative approach to assign data points to clusters and update the cluster centers until convergence.
It is an iterative clustering algorithm. The steps given below need to be followed for this algorithm:

Here's an example of how to apply K-means clustering using the KMeans class from scikit-learn:

**Step 1**: We need to specify the desired number of K subgroups.

**Step 2**: Fix the number of clusters and randomly assign each data point to a cluster. Or in other words we need to classify our data based on the number of clusters.

In this step, cluster centroids should be computed.

As this is an iterative algorithm, we need to update the locations of K centroids with every iteration until we find the global optima or in other words the centroids reach at their optimal locations.
The following code will help in implementing K-means clustering algorithm in Python. We are going to use the Scikit-learn module.

Let us import the necessary packages:

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
```

The following line of code will help in generating the two-dimensional dataset, containing four blobs, by using `make_blob` from the `sklearn.dataset` package.

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.40, random_state=0)
```

We can visualize the dataset by using the following code:

```python
plt.scatter(X[:, 0], X[:, 1], s=50);
plt.show()
```

Here, we are initializing kmeans to be the KMeans algorithm, with the required parameter
of how many clusters (n_clusters).

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
```

In the code above, we first create an instance of the KMeans class with the desired number of clusters (n_clusters). We then fit the model to the data using the `fit()` method. After fitting, we can use the predict() method to assign cluster labels to new data points. Finally, we retrieve the coordinates of the cluster centers using the cluster`_centers_` attribute.

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```
