## Converting the Model for Deployment on Amazon SageMaker

#### 1. Converting the Model:

Before deploying a model on Amazon SageMaker, it's essential to ensure that the model is in a format compatible with the platform. SageMaker supports various model formats, including TensorFlow's SavedModel, Apache MXNet's model archive (MAR), and PyTorch's TorchScript format.

Let's take a look at how you can convert models from popular frameworks into formats compatible with SageMaker:

##### a. TensorFlow Model:

````python
import tensorflow as tf

# Convert TensorFlow model to SavedModel format
model = tf.keras.models.load_model('your_model.h5')
tf.saved_model.save(model, 'tensorflow_model')
````

##### b. PyTorch Model:


````python
import torch

# Convert PyTorch model to TorchScript format
model = torch.load('your_model.pth')
model.eval()
torch.jit.save(torch.jit.trace(model, torch.randn(1, input_size)), 'pytorch_model.pt')
````



#### 2. Packaging the Model and Dependencies:

Amazon SageMaker requires that your model and any necessary dependencies are packaged into a compressed archive file (`.tar.gz`). This archive file is then uploaded to Amazon S3, from where SageMaker can access it during deployment.

Let's package the model and dependencies into a `.tar.gz` file:

```bash
# Create a directory to hold your model files and dependencies
mkdir model

# Copy your converted model file(s) and any necessary dependencies into the directory
cp tensorflow_model/* model/
# or
cp pytorch_model.pt model/

# Create a requirements.txt file containing any Python dependencies
echo "numpy==1.21.5" > model/requirements.txt

# Compress the directory into a .tar.gz file
tar -czvf model.tar.gz model
```

Once you have the `model.tar.gz` file, you can upload it to an Amazon S3 bucket. SageMaker will use this file during model deployment.


# Session 3: Training the Model on Amazon SageMaker

Now we'll dive into the process of training your machine learning model on Amazon SageMaker.

## 1. Configuring Training Instances and Hyperparameters:

When training a model on Amazon SageMaker, you have the flexibility to choose the type and number of instances for training, as well as configure hyperparameters for optimal model performance.

**Instance Types**: Amazon SageMaker offers a range of instance types optimized for different use cases, such as CPU-based instances for general-purpose tasks or GPU-based instances for deep learning workloads.

**Hyperparameters**: These are parameters that govern the training process and affect the performance and behavior of the trained model. Examples include learning rate, batch size, and regularization strength.

Below are steops on how to Configure for training:

**Step 1: Navigate to Amazon SageMaker Console**

- Begin by navigating to the Amazon SageMaker console. You can find this by searching for "SageMaker" in the AWS Management Console search bar and selecting the appropriate service.
image.png
**Step 2: Select Training Jobs**

- In the SageMaker console dashboard, locate the "Training jobs" under the Training Tab (dropdown) section in the left navigation pane and click on it.

image.png

**Step 3: Create Training Job**

- Click on the "Create training job" button to start configuring a new training job.

image.png


**Step 4: Specify Training Job Settings**

- Enter a unique name for your training job in the "Training job name" field.
- Select the IAM role that SageMaker will assume to perform tasks on your behalf.
- Choose the instance type for training. SageMaker offers a variety of instance types optimized for different workloads, such as CPU, GPU, and inference.
image.png
- Specify the number of instances and the instance type for training. You can choose from a range of instance types depending on your workload and budget.

**Step 5: Define Hyperparameters**

- Scroll down to the "Hyperparameters" section.
- Define hyperparameters specific to your machine learning algorithm or custom training script. These parameters govern the training process and can significantly impact model performance.
- You can specify hyperparameters as key-value pairs, such as learning rate, batch size, and regularization strength.

image.png

**Step 6: Configure Input & Output Location**

- Optionally, configure the input and output location where SageMaker will store the trained model artifacts and other input and output files.
- You can choose to store output in Amazon S3 or allow SageMaker to automatically create a new S3 bucket for you.

**Step 7: Additional Configuration (Optional)**

- Depending on your specific requirements, you may need to configure additional settings such as VPC, security settings, and data sources.

**Step 8: Review and Launch**

- Review the configuration settings to ensure everything is correct.
- Click on the "Create training job" button to launch the training job.

Amazon SageMaker supports both built-in algorithms and custom training scripts for model training.

**Built-in Algorithms**: SageMaker provides a collection of pre-built algorithms optimized for various machine learning tasks, such as linear regression, XGBoost, and image classification. You can simply specify the algorithm and hyperparameters, and SageMaker handles the rest.

# Session 4: Deploying the Model as an Endpoint

We'll explore the process of deploying your trained machine learning model as an endpoint on Amazon SageMaker. We'll cover creating an endpoint configuration, deploying the trained model, and testing the deployed model with sample data.

#### 1. Creating an Endpoint Configuration:

**Step 1: Navigate to Endpoints**

- Start by navigating to the Amazon SageMaker console and selecting "Endpoints" from the left navigation pane.

image.png

**Step 2: Create Endpoint Configuration**

- Click on the "**Create endpoint configuration**" button to create a new endpoint configuration.



**Step 3: Configure Endpoint**

- Enter a unique name for your endpoint configuration.
- Select the model that you want to deploy from the list of available models.
- Specify the instance type and number of instances for the endpoint. Choose an instance type that matches the requirements of your model and workload.
- Optionally, configure additional settings such as VPC, security settings, and data capture.

**Step 4: Create Endpoint Configuration**

- Review the configuration settings to ensure everything is correct.
- Click on the "Create endpoint configuration" button to create the endpoint configuration.

#### 2. Deploying the Trained Model as a SageMaker Endpoint:

**Step 1: Navigate to Endpoints**

- In the Amazon SageMaker console, select "Endpoints" from the left navigation pane.

**Step 2: Deploy Model**

- Click on the "Create endpoint" button to deploy the trained model as an endpoint.

**Step 3: Select Configuration**

- Choose the endpoint configuration that you created in the previous step.

**Step 4: Configure Endpoint**

- Enter a unique name for your endpoint.
- Review the instance type and number of instances configured in the endpoint configuration.

**Step 5: Deploy Endpoint**

- Click on the "Deploy endpoint" button to deploy the model as an endpoint.

#### 3. Testing the Deployed Model with Sample Data:

**Step 1: Obtain Endpoint URL**

- Once the endpoint is deployed successfully, you will receive an Endpoint URL.

**Step 2: Prepare Sample Data**

- Prepare sample data that you want to use for testing the deployed model. Ensure that the data format matches the input format expected by the model.

**Step 3: Invoke Endpoint**

- Use the SageMaker SDK or AWS CLI to invoke the endpoint with the sample data.

**Step 4: Evaluate Results**

- Analyze the results returned by the endpoint to verify that the model is producing the expected outputs.
